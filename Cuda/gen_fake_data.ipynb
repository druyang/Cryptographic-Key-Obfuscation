```python
import pandas as pd
import numpy as np
from random import randint
```

```python
patients = pd.read_csv('../Data/RAW_PatientInfo.csv')
patients['birth_year'] = patients['birth_year'].astype('UInt64')
patients.head()
```

```python
corr_patients = patients.copy()
corr_patients.loc[corr_patients['sex'] == 'male', 'state'] = 'deceased'
corr_patients.loc[corr_patients['sex'] != 'male', 'state'] = 'released'
```

```python
corr_patients[corr_patients['sex'] == 'male'].head()
```

```python
corr_patients[corr_patients['sex'] != 'male'].head()
```

```python
def gen_random_state(state):
    states = {0: 'released', 1: 'isolated', 2: 'deceased'}
    return(states[randint(0, 2)])
```

```python
uncorr_patients = patients.copy()
uncorr_patients['state'] = uncorr_patients['state'].apply(gen_random_state)
```

```python
uncorr_patients.head()
```

```python
corr_patients.to_csv('../Data/hypercorrelated.csv', index=False)
```

```python
uncorr_patients.to_csv('../Data/uncorrelated.csv', index=False)
```
